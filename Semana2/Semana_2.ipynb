## Medidas de Forma
------
Son aquellas que se pueden apreciar o **ver** cuando graficamos un **dataset**

### Curtosis

Esta medida de forma nos indica que tan **achantada** ó **escarpada** se encuentra una curva de una distribución, entre la **curtosis** sea de mayor tamaño significa que los datos se **encuentran más cerca a la media**.

* Curtosis > 0 -> Muy concentrados en la media
* Curtosis = 0 -> Distribución normal
* Curtosis < 0 -> Poco concentrados en la media

```python
# Creación de Series para el promedio

promedio = pd.Series(data.mean(numeric_only=True))

# Creación de Series para la curtosis

kurtosis = pd.Series(data.kurt(numeric_only = True))

# Merge de dataFrame promedio y kurtosis

prom_kur = pd.DataFrame(data = [promedio,kurtosis],index = ['Mean','Kurtosis']).T
prom_kur
```

# Medidas multivariadas

Es aquella parte de la **estadística descriptiva** que se encarga de **estudiar** los datos de una manera en conjunto dadas unas **variables especificas** puesto que en ocasiones puede que no exista un buen entendimiento si solamente se entienden estos datos de manera no relacionada con otros ( **análisis multivariado** ). Existen muchas técnicas y conceptos de esta rama, sin embargo dos de los más importantes son : 

- **Coeficiente de correlación**
    
    También conocido como ***coeficiente de relación de Pearson**,* nos indica de una manera numérica que tan relacionados se encuentran dos grupos de datos de **diferentes variables**. Este coeficiente oscila entre **-1 y 1,** en donde : 
    
    - Si el **coeficiente** se acerca al número **-1** se entiende que la ***relación*** entre los datos es de sentido ***inverso***, es decir, que a medida que uno aumenta el otro disminuye.
    - Si el **coeficiente** es igual a **0** se entiende que la ***no existe alguna relación.***
    - Si el **coeficiente** se acerca a **+1** se entiende que la ***relación*** entre los datos es de sentido ***positivo***, es decir, que a medida que uno aumenta el otro de igual manera.
    
    Es importante destacar que: 
    
    - Entre ***más grande*** sea la muestra el coeficiente será mucho ***más exacto***.
    - Si existen valores atípicos es mejor obviarlos puesto que estos pueden interferir en nuestra constante numérica.
    - Es mejor usar el ***coeficiente de Pearson*** en **dataset** que tengan una **relación lineal.**
        
        ![Untitled](https://s3-us-west-2.amazonaws.com/secure.notion-static.com/47127cf1-5a26-4c28-a5eb-a86f8cf9d7da/Untitled.png)
        
- **Covarianza**
    
    Se entiende como covarianza a aquella relación que existe entre **dos conjuntos de datos** y sus respectivas **medias**. Al igual que en el **coeficiente de correlación** existen tres grandes momentos los cuales son : 
    
    - Si la ***covarianza*** es menor a **0** se dice que hay una relación ***negativa.*** Alguna variable “sube” mientras la otra “baja”.
    - Si la ***covarianza*** es igual a **0** se dice que no hay alguna relación.
    - Si la ***covarianza*** es mayor que **0** se dice que existe una relación ***positiva***. Ambas variables “suben”.
    
    Es importante destacar que la **covarianza** unicamente importa el **signo** más no el **número** como en el ***coeficiente de correlación*** que determina no solamente el signo sino la *proporción de que tan relacionados* se encuentran los datos.
    
- **Tablas de contingencia o tablas de doble entrada**
    
    Son tablas usadas comúnmente para representar la relación que existe entre varios tipos de variables.

# Visualización de datos
----
- **Histograma**
    
    Es la representación de como se ***distribuyen*** las variables en **intervalos discretos** ( *intervalos de tamaño fijo ).* Este tipo de gráfica nos sirve para hallar ***tendencias, valores atípicos y representar la distribución que tienen el conjunto de datos.***
    
   
    
- **Gráfica de lineas**
    
    Es el tipo de gráfica que nos permite entender como se comporta respecto a otra a lo largo del tiempo, puesto que permite encontrar tendencias y se pueden aplicar conceptos de estadística inferencial para poder realizar **suposiciones.**
    
    
- **Gráfica de dispersión o scatter plot**
    
    Es un tipo de gráfica usada para representar **dos conjuntos de datos** y sirve para hallar la relación que estos pueden tener. 
  
    
- **Gráfica de barras**
    
    Es la gráfica que representa o muestra la frecuencia que tienen un grupo de datos provenientes de unas variables fijas, sirve para poder realizar comparaciones entre variables y sus frecuencias respectivas.
    
    
- **Diagrama de cajas**
    
    Es el tipo de gráfica que ilustra las **medidas de posición** como los cuartiles, media, percentiles, etc..
    
    
- **Gráfica circular**
    
    Es el tipo de gráfica que representa en términos de porcentaje la frecuencia que tienen un conjunto de datos específicos.
    
- **Gráfica hexagonal**
    
    Es una gráfica similar a la de barras por que permite ver la **frecuencia** de un conjunto de datos, sin embargo, es importante destacar que este tipo de grafica se diferencia ya que muestra la frecuencia en términos de dos variables diferentes.
